```python
orders['item_price']=orders.item_price.str.replace('$', '').astype(float)
```

```python
orders['item_price']
```

```python
orders.head(2)
```

```python
top_items = orders.groupby('item_name').quantity.sum().sort_values(ascending = False).head(4)
```

```python
my_top_items=list(top_items.index)
```

```python
orders_top=orders[orders.item_name.isin(my_top_items)]
```

```python
orders.columns
```

```python
orders.groupby('item_name').item_price.sum().sort_values(ascending=False).head(4)
```

```python
summed_orders = orders_top.groupby('item_name')\
.item_price.sum().reset_index()
```

```python
summed_orders
```

```python
sns.barplot(data=summed_orders, y='item_name', x='item_price')
plt.title('Most popular Chipotle items')
plt.ylabel('Item Name')
plt.xlabel('Summed price in USD')
plt.show()
```

```python
sleepers['subject_cat'] = 'Subject:' + sleepers.Subject.astype(str)
```

```python
plt.figure(figsize=(16,10))

sns.lineplot(data=sleepers, x ='Days', y='Reaction', hue='subject_cat', legend=False)


sns.lineplot(data=sleepers, x='Days', y='Reaction',color='black', legend=False)


plt.title('Reaction time of subject over time.')
plt.ylabel('Reaction time in ms')
plt.xlabel('days of sleep deprevation')
plt.annotate('Average of all subjects',(1,260))
plt.show()
```
