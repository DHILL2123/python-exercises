```python
x = np.linspace(-2,2,100)  
y = [n ** 2 for n in x]
```

```python
#Create and label 4 separate charts for the following equations (choose a range for x that makes sense):
# y=âˆšx
# y=x3
# y=2x
# y=1/(x+1)
# You can use functions from the math module to help implement some of the equations above
```

```python
# y=âˆšx
x = range(0,100)
y = [math.sqrt(n) for n in x]

plt.plot(x,y)

plt.title('$ y = \sqrt{x}$')
plt.xlabel('$x$')
plt.ylabel('$\sqrt{x}$')
plt.xlim(0,100)
plt.ylim(0, 10)

plt.grid(True, ls = '--');
plt.show()
```

```python
#y=x3
x1 = range(0, 20)
y1 = [n ** 3 for n in x1]

plt.plot(x1,y1)

plt.title('$ y = x^3$')
plt.xlabel('$x$')
plt.ylabel('$x^3$')
plt.xticks(range(0,20,5))

plt.xlim(0,20)
plt.ylim(0)
plt.grid(True, ls = '--');

plt.show;
```

```python
#y=2x
x2 = [x2 * 0.1 for x2 in range(0,150)]
y2 = [2 ** n for n in x2]

plt.plot(x2,y2)

plt.title('$y = 2^x$')
plt.xlabel('$x$')
plt.ylabel('$2^x$')

plt.show;
```
