3.Load the swiss dataset and read it's documentation. The swiss dataset is available from 
pydatset rather than seaborn. Create visualizations to answer the following questions:

```python
data('swiss')
dfswiss= data('swiss')
```

Create an attribute named is_catholic that holds a boolean value of whether or not the province is Catholic. 
(Choose a cutoff point for what constitutes catholic)

```python
dfswiss['is_catholic'] = dfswiss['Catholic'] >= 51
```

Does whether or not a province is Catholic influence fertility?

```python
#yes, fertility is higher in catholics
sns.barplot(data=dfswiss, x="is_catholic", y='Fertility')
plt.show()
```

What measure correlates most strongly with fertility?

```python
#agriculture
sns.scatterplot(data=dfswiss, x='Fertility', y='Agriculture')
plt.show()
```
